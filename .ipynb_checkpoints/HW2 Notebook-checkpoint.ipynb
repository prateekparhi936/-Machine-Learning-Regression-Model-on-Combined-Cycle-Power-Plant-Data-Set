```python
## 1(c)

from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
from statsmodels.api import OLS
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

lin_reg_coeff=np.empty([4])
columns=['AT','V','AP','RH']
i=0

for i,col in enumerate(columns):
    print("Linear Regression where X=",col,"and Y=PE")
    X=Data[[col]]
    Y=Data[['PE']]
    X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
    Lin_regression=LinearRegression().fit(X_train,Y_train)
    print("Linear Regression Score: ",Lin_regression.score(X_test,Y_test))
    print("Linear Regression Coeff: ",Lin_regression.coef_)
    lin_reg_coeff[i]=Lin_regression.coef_
    Y_predict=Lin_regression.predict(X_test)
    X_test= sm.add_constant(X_test)
    print("Summary of OLS : \n", OLS(Y_test,X_test).fit().summary())
    print("\n")
 
print("\nLinear Regression Coeff: ",lin_reg_coeff,"\n\n")
print("All the models are statistically significant with a p value less than 0.0001")
print("There are some outliers which exists in this dataset ")
print("\n\n")


## the plot below shows the relation between predictors and response
column_names=['AT','V','AP','RH']
handle=plt.figure(figsize=(22,22)) 
handle.subplots_adjust(hspace=0.3,wspace=0.3) ## spacing between height and width
for i,col_name in enumerate(column_names):
    handle.add_subplot(2,2,i+1)  ## (2 rows,3 cols) subplots and i starts from 0 so its i+1
    sn.regplot(y=col_name,x='PE',data=Data,marker='+',color='green')
    
    
```

```python
## 1(c)

print("\n----------------------** AFTER REMOVING OUTLIERS **-------------------\n")
    
## LINEAR REGRESSION AFTER REMOVING OUTLIER

z = np.abs(stats.zscore(Data))
Data = Data[(z < 2).all(axis=1)]
lin_reg_coeff=np.empty([4])
columns=['AT','V','AP','RH']

i=0

for i,col in enumerate(columns):
    print("Linear Regression where X=",col,"and Y=PE")
    X=Data[[col]]
    Y=Data[['PE']]
    X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
    Lin_regression=LinearRegression().fit(X_train,Y_train)
    print("Linear Regression Score: ",Lin_regression.score(X_test,Y_test))
    print("Linear Regression Coeff: ",Lin_regression.coef_)
    lin_reg_coeff[i]=Lin_regression.coef_
    Y_predict=Lin_regression.predict(X_test)
    X_test= sm.add_constant(X_test)
    print("Summary of OLS : \n", OLS(Y_test,X_test).fit().summary())
    print("\n")
 
print("\nLinear Regression Coeff: ",lin_reg_coeff,"\n\n")
print("All the models are statistically significant with a p value less than 0.0001")
print("There are some outliers which exists in this dataset ")
print("\n\n")

## the plot below shows the relation between predictors and response
column_names=['AT','V','AP','RH']
handle=plt.figure(figsize=(22,22)) 
handle.subplots_adjust(hspace=0.3,wspace=0.3) ## spacing between height and width
for i,col_name in enumerate(column_names):
    handle.add_subplot(2,2,i+1)  ## (2 rows,3 cols) subplots and i starts from 0 so its i+1
    sn.regplot(y=col_name,x='PE',data=Data,marker='+',color='green')
    
```

```python
## 1(i)

## KNN Regression using normalized feature
## find best k
## find train error
## find test error
## plot train and test error in terms of 1/k

from sklearn import preprocessing

DataNorm=Data
columns=['AT','V','AP','RH','PE']
DataNorm=preprocessing.normalize(DataNorm)
DataNorm=pd.DataFrame(DataNorm)
DataNorm.columns=columns

        
from sklearn.neighbors import KNeighborsRegressor
columns=['AT','V','AP','RH']
k_neighbors=np.arange(1,101,1)
maxScore_norm=0
most_suitable_k=1
train_error=np.empty([100])
test_error=np.empty([100])
X=DataNorm[columns]
Y=DataNorm['PE']
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,random_state=42,test_size=0.2)
for i,k in enumerate(k_neighbors):
    KNN_Reg=KNeighborsRegressor(n_neighbors=k,weights='uniform',p=2,algorithm='auto')
    KNN_Reg.fit(X_train,Y_train)
    Y_test_predict=KNN_Reg.predict(X_test)
    Y_train_predict=KNN_Reg.predict(X_train)
    score=KNN_Reg.score(X_test,Y_test)
    train_error[i]=(1-KNN_Reg.score(X_train,Y_train))*100
    test_error[i]=(1-KNN_Reg.score(X_test,Y_test))*100
    if(score>maxScore_norm):
        maxScore_norm=score
        most_suitable_k=k

print(most_suitable_k," ",maxScore_norm)


  
k_neighbors=[1/i for i in range(1,101,1)]

plt.title("Train vs Test error for normalized data")
plt.plot(k_neighbors,train_error,label='train error',color='blue')
plt.plot(k_neighbors,test_error,label='test error',color='red')
plt.legend()
plt.xlabel('1/k')
plt.ylabel('error')
plt.show()
```

```python
## 1(i)

## KNN Regression using raw features
## find best k
## find train error
## find test error
## plot train and test error in terms of 1/k

from sklearn.neighbors import KNeighborsRegressor

columns=['AT','V','AP','RH']
test_error=np.empty([100])
train_error=np.empty([100])
k_neighbors=np.arange(1,101,1)
most_suitable_k=0
X=Data[columns]
Y=Data['PE']
maxScore=0

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,random_state=42,test_size=0.2)
for i,k in enumerate(k_neighbors):
    KNN_Reg=KNeighborsRegressor(n_neighbors=k,p=2,weights='uniform',algorithm='auto')
    KNN_Reg.fit(X_train,Y_train)
    Y_predict=KNN_Reg.predict(X_test)
    score=KNN_Reg.score(X_test,Y_test)
    train_error[i]=(1-KNN_Reg.score(X_train,Y_train))*100
    test_error[i]=(1-KNN_Reg.score(X_test,Y_test))*100
    if(score>maxScore):
        maxScore=score
        most_suitable_k=k
        
print(most_suitable_k," ",score)

k_neighbors=[1/i for i in range(1,101,1)]

plt.title("train vs test error in terms of 1/k")
plt.plot(k_neighbors,train_error,color='blue',label='train_error')
plt.plot(k_neighbors,test_error,color='red',label='test_error')
plt.legend()
plt.xlabel('1/k')
plt.ylabel('errors')
plt.show()
```


**(2.4.1)
For each of parts (a) through (d), indicate whether we would generally
expect the performance of a flexible statistical learning method to be
better or worse than an inflexible method. Justify your answer.**

**(a) The sample size n is extremely large, and the number of predictors p is small.**
(ans) flexible method works better in this case. The increase in sample size although might increase the variance of
individual observation but overall variance around the sample mean and population mean decreases.

**(b) The number of predictors p is extremely large, and the number of observations n is small.**
(ans) Inflexible method works better in this case. Because the sample size is small the overall variance between sample mean
and population mean is high. So, using a flexible model may lead to overfitting. Hence we prefer an inflexible model 
over a flexible model

**(c) The relationship between the predictors and response is highly non-linear.**
(ans) As we have prior knowledge that the relation between predictor and response is highly non linear therefore 
a flexible model would be useful. Since, it can fit well as compared to an inflexible model which looks for somesort of linearity between predictors and response

**(d) The variance of the error terms, i.e. σ2 = Var(), is extremely high.**
(ans) When the variance of the error term is extremely high we prefer a infexible model because a flexible model may 
lead to overfitting because of the noise in the data.
